```python
x=np.linspace(0,20,100)
x
```

```python
plt.plot(np.sin(x))
```

```python
plt.plot(x,np.cos(x))
plt.plot(x,np.sin(x))
```

```python
plt.plot(np.cos(x))
plt.plot(np.sin(x))
```

```python
plt.plot(np.sin(x),'*--')
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(np.sin(x),'*-')
plt.subplot(1,2,2)
plt.plot(np.cos(x),'*--')
plt.show()
```

```python
plt.figure()
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(np.sin(x))
plt.subplot(2,2,2)
plt.plot(np.cos(x))
plt.subplot(2,2,3)
plt.plot(np.tan(x))
plt.subplot(2,2,4)
plt.plot(np.cos(x)+1)
plt.show()
```

```python
plt.figure()
plt.subplot(1,2,1)
plt.plot(np.sin(x),color='green')
plt.subplot(1,2,2)
plt.plot(np.cos(x),color='red')
plt.show()
```

```python
plt.plot(x,np.sin(x-0),color='g')
plt.plot(x,np.cos(x-1),color='r')
```

```python
plt.plot(x,np.sin(x-0),color='b')
plt.ylim(-1,5,0.75)
plt.xlim(1,15)
plt.axis("equal")
plt.grid(visible=True)
plt.title("basics of matplotlib")
plt.xlabel('values on x axis')
plt.ylabel('values in y axis')
plt.show()
```
